```python
import polars as pl
import plotly.express as px

from datetime import datetime
from meteostat import Hourly, Stations
```

```python
data = pl.read_csv("stream_images/data.csv")
data
```

```python
px.scatter(data, x="Datetime", y="Average Brightness", hover_name="Filename")
```

```python
# Create first scatter plot trace
fig = px.scatter(data.drop_nulls().filter(pl.col("Datetime") > "2024-09-18"), x="Datetime", y="Temperature")

meteostat_data = Hourly('74410', datetime(2024, 9, 18), datetime(2024,9,26)).fetch().reset_index()

# Add second scatter plot trace
fig.add_scatter(x=meteostat_data["time"], y=meteostat_data["temp_f"], mode='markers', name='Meteostat')

# Show combined plot
fig.show()
```
